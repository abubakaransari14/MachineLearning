# Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np

# Data
X = np.array([150, 155, 160, 165, 170, 175, 180, 185]).reshape(-1, 1)
y = np.array([50, 53, 55, 58, 65, 68, 70, 75])

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on test set
predictions = model.predict(X_test)
print("Predictions:", predictions)
print("Actual:", y_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("MSE:", mse)
print("R²:", r2)
```

Practice Question: Predict Car Prices

You have a dataset of used cars:

| Mileage (in 1000 km) | Price (in $1000) |
| -------------------- | ---------------- |
| 50                   | 20               |
| 60                   | 18               |
| 70                   | 17               |
| 80                   | 15               |
| 90                   | 14               |
| 100                  | 12               |


```python
carMileage = [50,60,70,80,90,100]
price = [20,18,17,15,14,12]
x = np.array(carMileage).reshape(-1,1)
y = np.array(price)

print(f"Type of X {type(x)} and {x.shape}")
print(f"Type of y {type(y)} and {y.shape}")

xTrain, xTest, yTrain, yTest = train_test_split(x,y, test_size=0.20, random_state=42)
model = LinearRegression()
model.fit(xTrain, yTrain)

prediction = model.predict(xTest)
print("Predicted:", prediction)
print("Actual:   ", yTest)

mse = mean_squared_error(yTest, prediction)
r2 = r2_score(yTest, prediction)

print("Mean Squared Error:", mse)
print("R² score:", r2)

new_car = np.array([[75]])
predicted_price = model.predict(new_car)
print("Predicted price for 75,000 km:", predicted_price[0])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))

# Training points
plt.scatter(xTrain, yTrain, color='blue', label='Training Data')

# Test points
plt.scatter(xTest, yTest, color='red', label='Test Data')

# Regression line (cover full range for smooth line)
x_line = np.linspace(45, 105, 100).reshape(-1,1)
y_line = model.predict(x_line)
plt.plot(x_line, y_line, color='green', label='Regression Line')

# New car prediction
plt.scatter(75, predicted_price, color='orange', marker='*', s=200, label='Prediction 75k km')

# Labels
plt.xlabel('Mileage (1000 km)')
plt.ylabel('Price ($1000)')
plt.title('Linear Regression: Car Price vs Mileage')
plt.legend()
plt.grid(True)
plt.show()
```
